```python
#box plot
plt.figure(figsize=(12, 10))
sns.boxplot(x=df_avg['Region'], y=df_avg['Life_expectancy'], data=df_avg, palette='Paired')
plt.grid(True)
plt.xticks(rotation=90);
```

```python
des = df_avg['Life_expectancy'].describe()
```

```python
des
```

```python
life_exp_less50 = df_avg[df_avg['Life_expectancy'] < des['mean']]
```

```python
life_exp_less50 
```

```python
#Bar chart 
plt.figure(figsize=(15, 6))
plt.bar(life_exp_less50['Country'], life_exp_less50['Life_expectancy'], color = '#f39c12')
plt.xlabel("Countries")
plt.ylabel("Life Expectancy")
plt.title("Countries with Life Expectancy Below the Mean")
plt.xticks(rotation=90)
plt.show()
```

```python
fig = px.choropleth(life_exp_less50, 
                    locations='Country', 
                    locationmode='country names', 
                    color='Life_expectancy', 
                    hover_name='Country', 
                    hover_data={'Region': True, 'Life_expectancy': True}, 
                    color_continuous_scale='Viridis', 
                    title='Average Life Expectancy by Country')

fig.show()
```

```python
top_10_life_exp = df_avg[['Country', 'Life_expectancy']].sort_values(by='Life_expectancy', ascending=False).head(10)
```

```python
#bar plot using sns
plt.figure(figsize=(10, 6))
sns.barplot(top_10_life_exp['Life_expectancy'], top_10_life_exp['Country'], palette='magma', orient = "h")
plt.xlabel("Countries")
plt.ylabel("Life Expectancy")
plt.title("Top 10 Countries by Life Expectancy ")
plt.show()
```

```python
#top 3 countries 
df_avg[['Country', 'Life_expectancy']].sort_values(by='Life_expectancy', ascending=False).head(3)
```

```python
#line Chart
selected_countries = ['Japan', 'Switzerland', 'Iceland']
df_selected = df[df['Country'].isin(selected_countries)]

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_selected, x='Year', y='Life_expectancy', hue='Country')
plt.title('Life Expectancy Over Time for Top 3 Countries')
plt.show()
```

```python
# Pair plot 
key_vars = ['Life_expectancy', 'GDP_per_capita', 'Alcohol_consumption', 'BMI', 'Incidents_HIV']
sns.pairplot(df_avg[key_vars], diag_kind='kde')
plt.suptitle('Pair Plot of Key Health Indicators', y=1.02)
plt.show()
```

```python
#histplot 

plt.figure(figsize=(10, 5))
sns.histplot(df_avg['Life_expectancy'], palette=sns.set_palette("viridis"), kde = True)
plt.xlabel("Life Expectancy")
plt.ylabel("Frequence")
plt.title("Distribution of Life Expectancy")
plt.show()
```

```python
#Scatter Plot 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_avg, x='GDP_per_capita', y='Life_expectancy', hue='Region', palette='viridis')
plt.title('GDP per Capita vs. Life Expectancy by Region')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.legend(title='Region')
plt.show()
```
